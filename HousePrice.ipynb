<font color="red" >
<h1 align="center"> USING RANDOM FOREST REGRESSOR </h1></font>
<br>
<h2> Applying Random Forest Regressor on the train set to train the model and then using the model to evaluate the test set</h2>
<BR>
<HR>

```python
from sklearn.ensemble import RandomForestRegressor
from sklearn.metrics import r2_score
import math
k=0
t1=0
max=0
c=0

for i in range(3,150):
    clf=RandomForestRegressor(n_estimators=i,random_state=3)
    clf.fit(trainx,trainy)
    yhat=clf.predict(testx)
    t1=r2_score(testy,yhat)
    if(t1>max):
        max=t1
        k=i
    i=i+1
print(max)
print(k)
rmse=math.sqrt(np.mean(((yhat-testy)**2)))
print(math.log(rmse))
```

```python
from sklearn import svm
from sklearn.metrics import r2_score
max=-100
i=0
l=0
for i in range(3,10):
    clf2=svm.SVR(kernel='poly',degree=i,gamma='auto')
    clf2.fit(trainx,trainy)
    yhat2=clf2.predict(testx)
    b=clf2.score(testx,testy)
    if(max<b):
        max=b
        l=i
    i=i+1
print(max)
print(l)
    
```
