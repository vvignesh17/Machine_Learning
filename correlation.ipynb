```python
%matplotlib inline
import numpy as np
from pylab import *
def de(x):
    xm=mean(x)
    return [xi-xm for xi in x]
def covariance(x,y):
    n=len(x)
    return dot(de(x),de(y))/(n-1)
pagespeed=np.random.normal(0.0,3.0,1000)
visited=np.random.normal(50.0,10.0,1000)
scatter(pagespeed,visited)
covariance(pagespeed,visited)

```

```python
scatter(pagespeed,visited/pagespeed)
covariance(pagespeed,visited/pagespeed)
```

```python
def corr(x,y):
    xs=x.std()
    ys=y.std()
    return covariance(x,y)/(xs*ys)
corr(pagespeed,visited/pagespeed)
```

```python
from numpy import random
random.seed(10)
totals={20:0,30:0,40:0,50:0,60:0,70:0}
purchase={20:0,30:0,40:0,50:0,60:0,70:0}
tp=0
for _ in range(10000):
    age=random.choice([20,30,40,50,60,70])
    pp=(float)(age)/100.0
    totals[age]+=1
    if(random.random()<pp):
        tp+=1
        purchase[age]+=1
```

```python
from numpy import random
random.seed(10)
totals={20:0,30:0,40:0,50:0,60:0,70:0}
purchase={20:0,30:0,40:0,50:0,60:0,70:0}
tp=0
for _ in range(10000):
    age=random.choice([20,30,40,50,60,70])
    #pp=40.0/100.0
    totals[age]+=1
    if(round(random.random()*10)%2==0):
        tp+=1
        purchase[age]+=1
```

```python
pe=(float)(tp)/10000.0
print("P(E)",pe)
```

```python
pef=(float)(purchase[40])/(float)(totals[40])
print("P(EUF)",pef)
```
