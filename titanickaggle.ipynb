```python
import pandas as pd
import matplotlib.pyplot as plt
from sklearn.metrics import accuracy_score
from sklearn.ensemble import RandomForestClassifier
from sklearn.tree import DecisionTreeClassifier
```

```python
df1=pd.read_csv("train.csv")
df1.head()
```

```python
d={"male":1,"female":0}
df1['Sex']=df1['Sex'].map(d)
d1={"S":2 ,"C":1,"G":0}
df1['Embarked']=df1['Embarked'].map(d1)
nan=df1.isnull().sum()
print(nan)
```

```python
na=nan[nan==0]
df1[na.keys()][:10]
```

```python
df1['Age'] = df1['Age'].fillna(0)
df1['Embarked']=df1['Embarked'].fillna(1)
df1.dtypes
nan=df1.isnull().sum()
print(nan)
```

```python
features=['Sex','Pclass','Age']
x=df1[features]
y=df1['Survived']
x.head()
```

```python
from sklearn.model_selection import train_test_split
trainx,testx,trainy,testy=train_test_split(x,y,random_state=4,train_size=0.762)
```

```python
from sklearn.metrics import accuracy_score
m=0
c=0
i=3
for i in range(3,100):
    model=RandomForestClassifier(random_state=4,n_estimators=i)
    model.fit(trainx,trainy)
    yhat=model.predict(testx)
    k=accuracy_score(testy,yhat)
    if(m<k):
        m=k
        c=i
print(c)
print(m)
```

```python
model=RandomForestClassifier(random_state=4,n_estimators=50)
model.fit(trainx,trainy)
yhat=model.predict(testx)
print(yhat)
```

```python
from sklearn.metrics import accuracy_score
yhat=model.predict(testx)
accuracy_score(testy,yhat)
```
